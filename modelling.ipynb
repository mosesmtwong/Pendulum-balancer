```python
import sympy as smp
from sympy import pprint
from sympy.physics.mechanics import dynamicsymbols, init_vprinting
init_vprinting()
```

```python
class DIPC:
    def __init__(self):
        self.half = smp.Rational(1, 2)
        self.t, self.g, self.u = smp.symbols("t g u")
        self.m0, self.m1, self.m2 = smp.symbols("m_0 m_1 m_2")
        self.l0, self.l1, self.l2 = smp.symbols("l_0 l_1 l_2")
        self.L0, self.L1, self.L2 = smp.symbols("L_0 L_1 L_2")
        self.I0, self.I1, self.I2 = smp.symbols("I_0 I_1 I_2")
        self.theta0, self.theta1, self.theta2 = dynamicsymbols("theta_0 theta_1 theta_2")
        self.T0 = self.half * self.m0 * self.theta0.diff(self.t) ** 2
        self.T1 = (self.half*self.m1*self.theta0.diff(self.t) ** 2+
                self.half*(self.m1*self.l1**2+self.I1)*self.theta1.diff(self.t)**2+
                self.m1*self.l1*self.theta0.diff(self.t)*self.theta1.diff(self.t)*smp.cos(self.theta1))
        self.T2 = (self.half*self.m2*self.theta0.diff(self.t)**2+
                self.half*self.m2*self.L1**2*self.theta1.diff(self.t))
                

pendulum = DIPC()
```

```python
class DIPC:
    def __init__(self):
        self.half = smp.Rational(1, 2)
        self.t, self.g, self.u = smp.symbols("t g u")
        self.m0, self.m1, self.m2 = smp.symbols("m_0 m_1 m_2")
        self.L1, self.L2 = smp.symbols("L_1 L_2")
        self.l1, self.l2 = (self.L1 / 2, self.L2 / 2)
        self.I1, self.I2 = smp.symbols("I_1 I_2")
        self.theta0, self.theta1, self.theta2 = smp.symbols("theta_0 theta_1 theta_2")
        self.w0, self.w1, self.w2 = smp.symbols(r"\dot{\theta_0} \dot{\theta_1} \dot{\theta_2}")
        self.a0, self.a1, self.a2 = smp.symbols('\u03B8\u0308_0 \u03B8\u0308_1 \u03B8\u0308_2')
        self.expr0 = ((self.m0+self.m1+self.m2)*self.a0 + (self.m1*self.l1+self.m2*self.L1)*smp.cos(self.theta1)*self.a1
                    + self.m2*self.l2*smp.cos(self.theta2)*self.a2 - (self.m1*self.l1+self.m2*self.L1)*smp.sin(self.theta1)*self.w1**2
                    - self.m2*self.l2*smp.sin(self.theta2)*self.w2**2 - self.u) 
        self.expr1 = ((self.m1*self.l1+self.m2*self.L1)*smp.cos(self.theta1)*self.a0
                    + (self.m2*self.l1**2+self.m2*self.L1**2+self.I1)*self.a1
                    + self.m2*self.L1*self.l2*smp.cos(self.theta1-self.theta2)*self.a2
                    + self.m2*self.L1*self.l2*smp.sin(self.theta1-self.theta2)*self.w2**2
                    - (self.m1*self.l1+self.m2*self.L1)*self.g*smp.sin(self.theta1))
        self.expr2 = (self.m2*self.l2*smp.cos(self.theta2)*self.a0 + self.m2*self.L1*self.l2*smp.cos(self.theta1-self.theta2)*self.a1
                    + (self.m2*self.l2**2+self.I2)*self.a2 - self.m2*self.L1*self.l2*smp.sin(self.theta1-self.theta2)*self.w1**2
                    - self.m2*self.l2*self.g*smp.sin(self.theta2))
        self.sols = smp.solve(
            [self.expr0, self.expr1, self.expr2], [self.a0, self.a1, self.a2]
        )

pendulum = DIPC()
pendulum.sols
```
